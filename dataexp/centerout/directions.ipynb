Plot vectors at specified angles with equal length.

```python
import numpy as np
import matplotlib.pyplot as plt

angles_deg = [0, 135, 180, 225, 270, 315, 45, 90]
length = 1.0

x = [length * np.cos(np.deg2rad(angle)) for angle in angles_deg]
y = [length * np.sin(np.deg2rad(angle)) for angle in angles_deg]

fig, ax = plt.subplots()
for angle, xi, yi in zip(angles_deg, x, y):
    ax.arrow(0, 0, xi, yi, head_width=0.05, head_length=0.1, fc="black", ec="black")
    xoff = 0.08 if xi == 0 else np.sign(xi) * 0.08
    yoff = 0.08 if yi == 0 else np.sign(yi) * 0.08
    ax.text(xi + xoff, yi + yoff, f"{angle}°", color="black", ha="center", va="center")

ax.set_aspect("equal", adjustable="box")
ax.spines["left"].set_color("black")
ax.spines["bottom"].set_color("black")
ax.spines["right"].set_color("none")
ax.spines["top"].set_color("none")
ax.xaxis.set_ticks([])
ax.yaxis.set_ticks([])
ax.set_xlim(-1.2, 1.2)
ax.set_ylim(-1.2, 1.2)
plt.show()
```
